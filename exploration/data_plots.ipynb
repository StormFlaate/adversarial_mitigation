```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Initial data
data_before = [('MEL', 1113), ('NV', 6705), ('BCC', 514), ('AKIEC', 327), ('BKL', 1099), ('DF', 115), ('VASC', 142)]
data_after = [('MEL', 3000), ('NV', 6705), ('BCC', 3000), ('AKIEC', 3000), ('BKL', 3000), ('DF', 3000), ('VASC', 3000)]

# Convert to DataFrame
df_before = pd.DataFrame(data_before, columns=['Skin lesion', 'Image count'])
df_before['Time'] = 'ISIC 2018 dataset'
df_after = pd.DataFrame(data_after, columns=['Skin lesion', 'Image count'])
df_after['Time'] = 'Data augmented dataset'

df = pd.concat([df_before, df_after])

# Create barplot
plt.figure(figsize=(10,6))
sns.barplot(x='Skin lesion', y='Image count', hue='Time', data=df)#, palette="Blues")
plt.title('ISIC 2018 dataset before and after data augmentation')
plt.savefig("isic_2018_dataset.png")
```
